## Lectura fitxer + samplejar

```python
def lectura_dataset(path, shuffle=False):
    dataset = pd.read_csv(path, header=0, delimiter=',')
    if shuffle:
        dataset = dataset.sample(frac=1).reset_index(drop=True)
    return dataset
```
